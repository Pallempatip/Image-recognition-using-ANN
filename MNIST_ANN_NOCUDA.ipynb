```python
import torch
import torch.nn as nn
import torchvision.transforms as tarnsforms
import torchvision.datasets as dataset
import matplotlib.pyplot as plt
from torch.autograd import variable
```

```python
train_dataset=dataset.MNIST(root="./data",train= True, transform=tarnsforms.ToTensor(),download= True)
test_dataset=dataset.MNIST(root="./data",train= False, transform=tarnsforms.ToTensor(),download= True)
```

```python
train_Load=torch.utils.data.DataLoader(dataset=train_dataset,shuffle= True,batch_size=100)
test_Load=torch.utils.data.DataLoader(dataset=test_dataset,shuffle= False,batch_size=100)
```

```python
class net(nn.Module):
    def __init__(self,input_Layer,Hidden_layer,Output_Layer):
        super(net,self).__init__()
        self.hidden=nn.Linear(input_Layer,Hidden_layer)
        self.activation=nn.ReLU()
        self.output=nn.Linear(Hidden_layer,Output_Layer)
    def forward(self,x):
        x= self.hidden(x)
        x=self.activation(x)
        x=self.output(x)
        return x
    
        
```

```python
network=net(input_Layer=784,Hidden_layer=400,Output_Layer=10)

optimizer= torch.optim.SGD(network.parameters(),lr=0.002)
Loss=nn.CrossEntropyLoss()
```

```python
for i,(images,Label) in enumerate(train_Load):
    print(images.size())
    images=images.view(-1,784)
    print(images.size())
```

```python
correct_train=0
total_train=0
for epochs in range(10):
    for i,(images,labels) in enumerate(train_Load):
        images= torch.tensor(images.view(-1,28*28))
        labels= torch.tensor(labels)
        optimizer.zero_grad()
        outputs=network(images)
        _,predicted= torch.max(outputs.data,1)
        
        total_train += labels.size(0)
        correct_train+=(predicted==labels).sum()
        
        loss=Loss(outputs,labels)
        loss.backward()
        optimizer.step()
        if(i+1)% 100==0:
            print('Epoch [{}/{}], Iteration [{}/{}], Training Loss: {}, Training Accuracy: {}%'.format
                 (epochs+1, epochs, i+1, len(train_dataset)//100, loss.item(), (100*correct_train/total_train)))
print("Done Training!")
```

```python
#test the network with test data
total=0
correct=0
for images,labels in test_Load:
    images= torch.tensor(images.view(-1,28*28))
    labels= torch.tensor(labels)
    out=network(images)
    _,predictes=torch.max(out.data,1)
    total +=images.size(0)
    correct +=(predictes==labels).sum()
Test_accuracy=100*correct/total
print(Test_accuracy.item())
    
    
```
